3. **Mode**

```python
import statistics as st
print(f"The mode salary is: {st.mode(df['TotalPay'])}")

#or
X = df['TotalPay']
def MaxMode(list_):

    freq = {}
    for i in list_:

        freq.setdefault(i, 0)
        freq[i] += 1
    highestfreq = max(freq.values())

    for key, value in freq.items():
        if value == highestfreq:
            return key

        
print(f"The mode salary is: {MaxMode(X)}")

#or
Mod = df[['TotalPay']].value_counts()
res = Mod.to_frame().reset_index()
print(f"The mode salary is: {res['TotalPay'][0]}")
```

2. **Represent the proportion of employees in different departments**

As we can see above, we have too many different JobTitles and not all of them mentioned the department.
So I'll take the data which have "DEPARTMENT" word on it, and check them manualy.

```python
Jobs = df["JobTitle"]
InDepartment = df["JobTitle"].str.upper().str.contains("DEPARTMENT")
for i in Jobs[InDepartment].unique():
    print(i)
```

As we can see here, there is two department "Police and Fire" and maybe three with "MEDICAL STAFF SERVICES DEPARTMENT".

I will group the small sections of the pie charts together under name "Others" to avoid making the chart hard to understand.

```python
Jobs = df["JobTitle"]
InFireDepartment = Jobs.str.upper().str.contains("FIRE DEPARTMENT")
FireDepartmentJobs = Jobs[InFireDepartment].unique()
iter_list_F = []
labels_F = []
colors_F = ("cyan", "indigo", "skyblue")
for i in range(len(FireDepartmentJobs)):
    val = sum(Jobs.str.upper().str.contains(FireDepartmentJobs[i].upper(), regex=False))
    if round((val*100)/sum(InFireDepartment),1)>=5:
        iter_list_F.append(val)
        labels_F.append(FireDepartmentJobs[i] + " " + str(round((val*100)/sum(InFireDepartment), 1)) + "%")

if sum(iter_list_F) != sum(InFireDepartment):
    val = sum(InFireDepartment)-sum(iter_list_F)
    iter_list_F.append(val)
    labels_F.append(f"Others {round((100*val)/sum(InFireDepartment),1)}%")
        
patches_F, texts = plt.pie(iter_list_F, colors = colors_F)
plt.legend(patches_F, labels_F, loc="center left", bbox_to_anchor=(1, 0, 0.5, 1), title = f"Job title(Total: {sum(InFireDepartment)})", title_fontsize = 16, fontsize = 14)
plt.title("Proportion of employees in \nFIRE department", fontsize = 18)
plt.show()
```

```python
Jobs = df["JobTitle"]
InPoliceDepartment = Jobs.str.upper().str.contains("POLICE DEPARTMENT")
PoliceDepartmentJobs = Jobs[InPoliceDepartment].unique()
iter_list_P = []
labels_P = []
colors_P = ("springgreen", "aquamarine", "lightslategray", "darkturquoise", "powderblue", "lightskyblue")
for i in range(len(PoliceDepartmentJobs)):
    val = sum(Jobs.str.upper().str.contains(PoliceDepartmentJobs[i].upper(), regex=False))
    if round((val*100)/sum(InPoliceDepartment),1)>=3:
        iter_list_P.append(val)
        labels_P.append(PoliceDepartmentJobs[i] + " " + str(round((val*100)/sum(InPoliceDepartment), 1)) + "%")

if sum(iter_list_P) != sum(InPoliceDepartment):
    val = sum(InPoliceDepartment)-sum(iter_list_P)
    iter_list_P.append(val)
    labels_P.append(f"Others {round((100*val)/sum(InPoliceDepartment),1)}%")
    
patches_P, texts = plt.pie(iter_list_P, colors = colors_P)
plt.legend(patches_P, labels_P, loc="center left", bbox_to_anchor=(1, 0, 0.5, 1), title = f"Job title(Total: {sum(InPoliceDepartment)})", title_fontsize = 16, fontsize = 14)
plt.title("Proportion of employees in \nPOLICE department", fontsize = 18)
plt.show()
```

```python
Jobs = df["JobTitle"]
InMSSDepartment = Jobs.str.upper().str.contains("MEDICAL STAFF SERVICES DEPARTMENT")
MSSDepartmentJobs = Jobs[InMSSDepartment].unique()
iter_list_M = []
labels_M = []
colors_M = ("deepskyblue", "limegreen")
for i in range(len(MSSDepartmentJobs)):
    val = sum(Jobs.str.upper().str.contains(MSSDepartmentJobs[i].upper(), regex=False))
    if round((val*100)/sum(InMSSDepartment),1)>=3:
        iter_list_M.append(val)
        MSSDepartmentJobs[i] = "MMS " + MSSDepartmentJobs[i].replace("MEDICAL STAFF SERVICES",'')
        labels_M.append(MSSDepartmentJobs[i] + " " + str(round((val*100)/sum(InMSSDepartment), 1)) + "%")
    
patches_M, texts = plt.pie(iter_list_M, colors = colors_M)
plt.legend(patches_M, labels_M, loc="center left", bbox_to_anchor=(1, 0, 0.5, 1), title = f"Job title(Total: {sum(InMSSDepartment)})", title_fontsize = 16, fontsize = 14)
plt.title("Proportion of employees in \nMEDICAL STAFF SERVICES department", fontsize = 18)
plt.show()
```

### 6:Simple Correlation Analysis:

```python
A = df[['BasePay','OvertimePay','OtherPay','Benefits','TotalPayBenefits','Year']]
print("The correlation between Salary and\n")
for i in A:
    print(f"{i} is: {df[i].corr(df['TotalPay'])}")
```

Here we noticed that the most correlation columns with salary are 1-TotalPayBenefits, 2-BasePay and 3-Benefits. I will show relationship between each one of them with the salary.

```python
x = df["TotalPayBenefits"]
y = df["TotalPay"]
df.plot(kind = "scatter", x = "TotalPayBenefits", y = "TotalPay")
plt.plot(np.unique(x), np.poly1d(np.polyfit(x, y, 1))(np.unique(x)), color='red')
plt.xlabel('TotalPayBenefits')
plt.ylabel('TotalPay')
plt.xlim(-650,568000)
plt.ylim(-650,568000)
plt.title("Correlation between TotalPayBenefits and TotalPay")
plt.grid(True)
plt.show()
```

```python
x = df["BasePay"]
y = df["TotalPay"]
df.plot(kind = "scatter", x = "BasePay", y = "TotalPay")
plt.plot(np.unique(x), np.poly1d(np.polyfit(x, y, 1))(np.unique(x)), color='red')
plt.xlabel('BasePay')
plt.ylabel('TotalPay')
plt.xlim(-10,167500)
plt.ylim(-650,568000)
plt.grid(True)
plt.title("Correlation between BasePay and TotalPay")
plt.show()
```
